## COVIDP Data

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
csv_file = "covid_p.csv"
covidp = pd.read_csv(csv_file)
covidp.head()
```

```python
covidp.columns
```

```python
covidp = covidp.drop(columns=['Province'])
covidp.head()
```

```python
#Data Analysis for US
us = covidp[covidp["Country"]=="US"]
us.head()
```

```python
us=us.fillna(0)
us.head()
```

```python
us.columns
```

```python
covidp_us=us.drop(columns='Continent')
covidp_us.head()
```

```python
covp_us=us.drop(columns='Country')
covp_us.head()
```

```python
# Export the cleaned data as a csv file
covidp_us.to_csv("covidp_us.csv")
 
```
